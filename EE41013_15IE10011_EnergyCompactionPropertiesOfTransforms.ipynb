```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
import math
```

```python
n = np.arange(0,64,1)
x = (0.5)**n
```

```python
x
```

```python
plt.plot(x)
```

```python
import cmath
def dft(N):
    W = np.ndarray((N,N),dtype = complex)
    for i in range(N):
        for j in range(N):
            W[i][j] = cmath.exp(complex(0,float(-2*math.pi/N*i*j)))
    return W
Y = dft(64)
```

```python
X = x.transpose()
z = np.matmul(Y,X)
```

```python
plt.plot(abs(z))
```

```python
def haarMatrix(n, normalized=True):
    n = 2**np.ceil(np.log2(n))
    if n > 2:
        h = haarMatrix(n / 2)
    else:
        return np.array([[1, 1], [1, -1]])

    # calculate upper haar part
    h_n = np.kron(h, [1, 1])
    # calculate lower haar part 
    if normalized:
        h_i = np.sqrt(n/2)*np.kron(np.eye(len(h)), [1, -1])
    else:
        h_i = np.kron(np.eye(len(h)), [1, -1])
    # combine parts
    h = np.vstack((h_n, h_i))
    return h
A = haarMatrix(64)
```

```python
b = np.matmul(A,X)
```

```python
plt.plot(b)
```

```python
from numpy.linalg import inv
El = []
for L in range(1,65):
    bm = []
    sum = 0
    for i in range(64-L):
        bm.append(b[i])
    for i in range(64-L,64):
        bm.append(0)
    xm = np.matmul(inv(A),bm)
    for i in range(len(X)):
        sum = sum + ((X[i]-xm[i])**2)
        q = sum/64
    El.append(q)
plt.plot(El)
```

```python
Dl = []
L = np.arange(1,64,1)
for j in range(len(L)):
    zm = []
    sum = 0
    for i in range(math.floor((65-L[j])/2)):
        zm.append(z[i])
    for i in range(math.floor((65-L[j])/2),math.ceil((65+L[j])/2)):
        zm.append(0)
    for i in range(math.ceil((65+L[j])/2),64):
        zm.append(z[i])
    dm = np.matmul(inv(Y),zm)
    for i in range(len(X)):
        sum = sum + ((abs(X[i]-dm[i]))**2)
        q = sum/64
    Dl.append(q)
```

```python
plt.plot(Dl)
```

```python
from scipy.fftpack import dct,idct
c = dct(X)
```

```python
Cl = []
for L in range(1,65):
    cm = []
    sum = 0
    for i in range(64-L):
        cm.append(c[i])
    for i in range(64-L,64):
        cm.append(0)
    xm = idct(cm)/128
    for i in range(len(X)):
        sum = sum + ((X[i]-xm[i])**2)
        q = sum/64
    Cl.append(q)
plt.plot(Cl)
```

```python
plt.plot(Dl)
plt.plot(El)
plt.plot(Cl)
plt.ylabel('E(L)')
plt.xlabel('L')
```
